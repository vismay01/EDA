```python
df.shape
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
# Renaming our columns
df = df.rename(columns={'coaster_name': 'Coaster_Name',
                   'year_introduced': 'Year_Introduced',
                   'opening_date_clean': 'Opening_Date',
                   'speed_mph': 'Speed_mph',
                   'height_ft': 'Height_ft',
                   'Inversions_clean': 'Inversions',
                   'Gforce_clean': 'Gforce'})
```

```python
df.isna().sum()
```

```python
df.loc[df.duplicated(subset=['Coaster_Name'])]
```

```python
# checking an example duplicate
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
df.columns
```

```python
# removing duplicates 
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])]
```

```python
#reset Index
df = df.reset_index(drop=True)
```

### Step 5: Ask a Question about the data
<ul><li>Try to answer a question you have about the data using a plot or static</li></ul>

What are the locations with the fastest with the fastest roller coasters (minimum of 10)?

```python
ax = df.query('Location != "Other"').groupby('Location')['Speed_mph'] \
    .agg(['mean','count']).query('count >= 10').sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12,5), title='Average Coaster Speed by Location')

ax.set_xlabel('Average Coaster Speed')
plt.show()
```
